# **Square Root Function**

```python
df = pd.read_csv('/content/covid19.csv')
df
```

```python
plt.scatter(df["confirmados_novos"], df["ativos"])
plt.show()
```

```python
df["conf_new_root"] = np.sqrt(df["confirmados_novos"])
```

```python
sns.distplot( a=df["confirmados_novos"], hist=True, kde=False, rug=False )
```

```python
sns.distplot( a=df["conf_new_root"], hist=True, kde=False, rug=False )
```

```python
plt.scatter(df["conf_new_root"], df["ativos"])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = df['ativos']
X1 = df['conf_new_root']
X1a = sm.add_constant(X1)
model_1a = sm.OLS(y, X1a)
```

```python
results_model_1a = model_1a.fit()
print(results_model_1a.summary())
```

From the obtained results we can build the equation
y = -4977.33+1275.46sqrt(x)

```python
import numpy as np
import matplotlib.pyplot as plt
from lmfit import Model
```

```python
def sinfunc(x, a, b, c, d):
    return a * np.sin(b*(x - c)) + d
```

```python
model  = Model(sinfunc)
```

```python
params = model.make_params(a=1000, b=0.1, c=0, d=0)
```

```python
#params['a'].min = 0
#params['c'].min = -np.pi
#params['c'].max = np.pi
```

```python
predictor = df["ID"]
predicted = df["confirmados_novos"]
result = model.fit(predicted, params, x=predictor)
```

```python
print(result.fit_report())
```

```python
plt.scatter(predictor,predicted,label='Confirmados_Novos')
plt.plot(predictor, result.best_fit, 'r-',label='Fitted function')
plt.title("ID vs Confirmados_Novos")
plt.xlabel('ID')
plt.ylabel('Confirmados_Novos')
plt.legend()
plt.show()
```
